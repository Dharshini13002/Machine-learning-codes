```python
import os
from PIL import Image
import numpy as np
from sklearn.metrics.pairwise import euclidean_distances

def load_images_from_folder(folder, resize_to=(100, 100)):
    images = []
    for filename in os.listdir(folder):
        img_path = os.path.join(folder, filename)
        if os.path.isfile(img_path):
            try:
                img = Image.open(img_path).convert('L').resize(resize_to)
                images.append(np.array(img).flatten())
            except Exception as e:
                print(f"Error loading image {filename}: {e}")
    return np.array(images)

folder_path = '/content/drive/MyDrive/sampleimagedb'
image_data = load_images_from_folder(folder_path)

if image_data.size > 0:
    distance_matrix = euclidean_distances(image_data)
    print("Euclidean Distance Matrix:")
    print(distance_matrix)

    if image_data.shape[0] >= 2:
        distance_first_second = euclidean_distances(image_data[0].reshape(1, -1), image_data[1].reshape(1, -1))
        print(f"\nEuclidean distance between first and second image: {distance_first_second[0][0]}")
else:
    print("No images loaded from the folder.")
```

```python
import os
from PIL import Image
import numpy as np
from sklearn.metrics.pairwise import euclidean_distances
import matplotlib.pyplot as plt

def load_images_from_folder(folder, resize_to=(100, 100)):
    images = []
    filenames = []
    for filename in os.listdir(folder):
        img_path = os.path.join(folder, filename)
        if os.path.isfile(img_path):
            try:
                img = Image.open(img_path).convert('L').resize(resize_to)
                images.append(np.array(img).flatten())
                filenames.append(filename)
            except Exception as e:
                print(f"Error loading image {filename}: {e}")
    return np.array(images), filenames

folder_path = '/content/drive/MyDrive/sampleimagedb'
image_data, filenames = load_images_from_folder(folder_path)

if image_data.size > 0:
    # Calculate Euclidean distances
    distance_matrix = euclidean_distances(image_data)
    print("Euclidean Distance Matrix:")
    print(distance_matrix)

    # Show the first two images side by side
    if len(filenames) >= 2:
        fig, axs = plt.subplots(1, 2, figsize=(8,4))

        # Reload images for display (not flattened)
        img1 = Image.open(os.path.join(folder_path, filenames[0])).convert('L').resize((100,100))
        img2 = Image.open(os.path.join(folder_path, filenames[1])).convert('L').resize((100,100))

        axs[0].imshow(img1, cmap='gray')
        axs[0].set_title(filenames[0])
        axs[0].axis('off')

        axs[1].imshow(img2, cmap='gray')
        axs[1].set_title(filenames[1])
        axs[1].axis('off')

        plt.show()

        # Print their Euclidean distance
        dist = distance_matrix[0,1]
        print(f"\nEuclidean distance between '{filenames[0]}' and '{filenames[1]}': {dist}")
else:
    print("No images loaded from the folder.")
```

```python
# prompt: for hamming code

# Simple Hamming distance calculation function
def hamming_distance(vec1, vec2):
    # Ensure vectors are of the same length
    if len(vec1) != len(vec2):
        raise ValueError("Vectors must have the same length")

    # Hamming distance is the number of positions at which the corresponding symbols are different
    return np.sum(vec1 != vec2)

if image_data.size > 0:
    # Convert image data to binary representation (e.g., thresholding)
    # A simple approach is to threshold at the mean value of the pixel data
    binary_image_data = (image_data > np.mean(image_data)).astype(int)

    # Calculate Hamming distances between all pairs of binary images
    num_images = binary_image_data.shape[0]
    hamming_dist_matrix = np.zeros((num_images, num_images))

    for i in range(num_images):
        for j in range(num_images):
            hamming_dist_matrix[i, j] = hamming_distance(binary_image_data[i], binary_image_data[j])

    print("\nHamming Distance Matrix (after binarization):")
    print(hamming_dist_matrix)

    # Print the Hamming distance between the first two binary images
    if len(filenames) >= 2:
        hamming_dist_first_second = hamming_dist_matrix[0, 1]
        print(f"\nHamming distance between '{filenames[0]}' and '{filenames[1]}' (after binarization): {hamming_dist_first_second}")
else:
    print("No images loaded from the folder.")
```

```python
# prompt: jaccard

# Function to calculate Jaccard similarity
def jaccard_similarity(vec1, vec2):
    # Ensure vectors are binary (0 or 1) for proper Jaccard calculation
    if not np.all(np.isin(vec1, [0, 1])) or not np.all(np.isin(vec2, [0, 1])):
        raise ValueError("Vectors must be binary (contain only 0 or 1)")

    # Calculate the intersection (number of positions where both are 1)
    intersection = np.sum(np.logical_and(vec1, vec2))

    # Calculate the union (number of positions where at least one is 1)
    union = np.sum(np.logical_or(vec1, vec2))

    # Avoid division by zero if the union is empty
    if union == 0:
        return 0.0  # Jaccard similarity is 0 if both sets are empty

    # Jaccard similarity is intersection / union
    return intersection / union

# Function to calculate Jaccard distance
def jaccard_distance(vec1, vec2):
    # Jaccard distance is 1 - Jaccard similarity
    return 1 - jaccard_similarity(vec1, vec2)

if image_data.size > 0:
    # Convert image data to binary representation (e.g., thresholding)
    # A simple approach is to threshold at the mean value of the pixel data
    binary_image_data = (image_data > np.mean(image_data)).astype(int)

    # Calculate Jaccard distances between all pairs of binary images
    num_images = binary_image_data.shape[0]
    jaccard_dist_matrix = np.zeros((num_images, num_images))

    for i in range(num_images):
        for j in range(num_images):
            # Use the jaccard_distance function
            jaccard_dist_matrix[i, j] = jaccard_distance(binary_image_data[i], binary_image_data[j])

    print("\nJaccard Distance Matrix (after binarization):")
    print(jaccard_dist_matrix)

    # Print the Jaccard distance between the first two binary images
    if len(filenames) >= 2:
        jaccard_dist_first_second = jaccard_dist_matrix[0, 1]
        print(f"\nJaccard distance between '{filenames[0]}' and '{filenames[1]}' (after binarization): {jaccard_dist_first_second}")

    # Optionally, calculate Jaccard similarity matrix as well
    jaccard_sim_matrix = np.zeros((num_images, num_images))
    for i in range(num_images):
        for j in range(num_images):
            jaccard_sim_matrix[i, j] = jaccard_similarity(binary_image_data[i], binary_image_data[j])

    print("\nJaccard Similarity Matrix (after binarization):")
    print(jaccard_sim_matrix)

    # Print the Jaccard similarity between the first two binary images
    if len(filenames) >= 2:
        jaccard_sim_first_second = jaccard_sim_matrix[0, 1]
        print(f"\nJaccard similarity between '{filenames[0]}' and '{filenames[1]}' (after binarization): {jaccard_sim_first_second}")

else:
    print("No images loaded from the folder.")
```
