```python
# Replace 'details.txt' with your uploaded file name
with open('/content/drive/MyDrive/smaplemlfile.txt', 'r') as file:
    content = file.read()

print(content)
```

```python
people = []
person = {}

for line in content.splitlines():
    if line.strip() == "":
        if person:
            people.append(person)
            person = {}
    else:
        key, value = line.split(":", 1)
        person[key.strip()] = value.strip()

# Add the last person if not already added
if person:
    people.append(person)

# Print structured data
for p in people:
    print(p)
```

```python
import pandas as pd

df = pd.DataFrame(people)
print(df)
```

```python
# Step 1: Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Step 2: Import Required Libraries
import os
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

# Step 3: Define the Image Directory Path
image_folder = "/content/drive/MyDrive/sampleimagedb"

# Step 4: List and Read Image Files
image_files = [f for f in os.listdir(image_folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

# Step 5: Display First Few Images
for i, img_name in enumerate(image_files[:5]):  # Show only first 5 images
    img_path = os.path.join(image_folder, img_name)
    image = Image.open(img_path)

    plt.imshow(image)
    plt.title(f"Image: {img_name}")
    plt.axis('off')
    plt.show()

# Step 6: Convert All Images to NumPy Arrays
image_data = []
for img_name in image_files:
    img_path = os.path.join(image_folder, img_name)
    try:
        image = Image.open(img_path).convert('RGB')  # Convert to RGB
        image = image.resize((128, 128))  # Resize to fixed size (128x128)
        image_array = np.array(image)
        image_data.append(image_array)
    except Exception as e:
        print(f"Error loading {img_name}: {e}")

print(f"\n Loaded {len(image_data)} images as NumPy arrays.")
print(f"Shape of first image: {image_data[0].shape}")
```

```python
# prompt: to read a code for xml file

# To read an XML file in Google Colab, you can use the `xml.etree.ElementTree` module
# which is part of Python's standard library, so no installation is needed.

import xml.etree.ElementTree as ET

# Assume you have an XML file named 'example.xml' in your current directory
# You can upload an XML file using files.upload() if it's not already there

# Example: Create a dummy XML file for demonstration
xml_content = """
<root>
    <person id="1">
        <name>Alice</name>
        <age>30</age>
    </person>
    <person id="2">
        <name>Bob</name>
        <age>25</age>
    </person>
</root>
"""
with open('example.xml', 'w') as f:
    f.write(xml_content)

# Replace 'example.xml' with the actual name of your XML file
xml_file_path = 'example.xml'

try:
    # Parse the XML file
    tree = ET.parse(xml_file_path)
    root = tree.getroot()

    # Now you can navigate and extract data from the XML tree
    print(f"Root tag: {root.tag}")

    # Iterate through elements
    for person in root.findall('person'):
        person_id = person.get('id')
        name = person.find('name').text
        age = person.find('age').text
        print(f"Person ID: {person_id}, Name: {name}, Age: {age}")

except FileNotFoundError:
    print(f"XML file not found at: {xml_file_path}")
except ET.ParseError as e:
    print(f"Error parsing XML file: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")

# You can also read XML from a string directly if needed
xml_string_data = """
<catalog>
   <book id="bk101">
      <author>Gambardella, Matthew</author>
      <title>XML Developer's Guide</title>
      <genre>Computer</genre>
      <price>44.95</price>
   </book>
</catalog>
"""

try:
    root_from_string = ET.fromstring(xml_string_data)
    print("\nReading from XML string:")
    for book in root_from_string.findall('book'):
        title = book.find('title').text
        author = book.find('author').text
        print(f"Book Title: {title}, Author: {author}")

except ET.ParseError as e:
    print(f"Error parsing XML string: {e}")
except Exception as e:
    print(f"An unexpected error occurred while reading string: {e}")
```

```python
# prompt: to read docx

!pip install python-docx

from docx import Document

# Replace 'your_document.docx' with the path to your DOCX file
# If the file is in your Drive, use the appropriate path after mounting
docx_file_path = '/content/drive/MyDrive/ccg feedback.docx' # Replace with your path

try:
    # Open the document
    document = Document(docx_file_path)

    print(f"Successfully opened document: {docx_file_path}")

    # Read paragraphs
    print("\n--- Reading Paragraphs ---")
    for paragraph in document.paragraphs:
        print(paragraph.text)

    # Read tables (if any)
    print("\n--- Reading Tables ---")
    for i, table in enumerate(document.tables):
        print(f"--- Table {i+1} ---")
        for row in table.rows:
            row_text = [cell.text for cell in row.cells]
            print("\t".join(row_text)) # Print cells separated by tab

except FileNotFoundError:
    print(f"Error: DOCX file not found at {docx_file_path}")
except Exception as e:
    print(f"An error occurred while reading the document: {e}")
```
