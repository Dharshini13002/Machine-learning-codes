```python
# prompt: barchart code inserting a file path

import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the specified file path
df = pd.read_csv('/content/sample_data/california_housing_train.csv')

# Example: Create a bar chart of the 'housing_median_age' column
# First, count the occurrences of each value
counts = df['housing_median_age'].value_counts().sort_index()

# Create the bar chart
plt.figure(figsize=(12, 6))
counts.plot(kind='bar')
plt.title('Distribution of Housing Median Age')
plt.xlabel('Housing Median Age')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# prompt: code for scatter plot in same way median income red color median house value blue color green and blue

# Scatter plot for median income (red) and median house value (blue)
plt.figure(figsize=(10, 6))

# Scatter plot for median income (red)
plt.scatter(df['latitude'], df['longitude'], c=df['median_income'], cmap='Reds', label='Median Income')

# Scatter plot for median house value (blue)
# We need another color map for house value, e.g., Blues or Greens
plt.scatter(df['latitude'], df['longitude'], c=df['median_house_value'], cmap='Blues', alpha=0.6, label='Median House Value')


plt.title('Scatter plot of Median Income and Median House Value')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.legend()
plt.colorbar(label='Value') # Add a color bar to indicate the values represented by colors
plt.show()
```

```python
# prompt: code for column chart

# Example: Create a column chart (similar to a bar chart) for the distribution of 'housing_median_age'
# First, count the occurrences of each value
counts = df['housing_median_age'].value_counts().sort_index()

# Create the column chart
plt.figure(figsize=(12, 6))
plt.bar(counts.index, counts.values)
plt.title('Distribution of Housing Median Age (Column Chart)')
plt.xlabel('Housing Median Age')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# prompt: /content/drive/MyDrive/titanic.csv    for this path give me  line chart for age till 30 only till age 30 with nonly 14 points

# Load the titanic data
df_titanic = pd.read_csv('/content/drive/MyDrive/titanic.csv')

# Filter for age till 30
df_age_30 = df_titanic[df_titanic['Age'] <= 30].copy()

# Drop rows where 'Age' is NaN before grouping
df_age_30.dropna(subset=['Age'], inplace=True)

# Group by age and count occurrences, then sort by age
age_counts = df_age_30['Age'].value_counts().sort_index()

# Select approximately 14 points by sampling the data
# We'll take every n-th point where n is roughly total points / 14
n = max(1, len(age_counts) // 14)
age_counts_sampled = age_counts.iloc[::n]


# Create the line chart
plt.figure(figsize=(10, 6))
plt.plot(age_counts_sampled.index, age_counts_sampled.values, marker='o', linestyle='-')
plt.title('Line Chart of Age Distribution (Up to 30)')
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid(True)
plt.tight_layout()
plt.show()
```
